**Scale Features**

When the data has different values, and even different measurement units, it can be difficult to compare them.


--------------------------------------------

so,we can scale data into new values that are easier to compare.

# The standardization method uses this formula:

z = (x - u) / s

Where z is the new value, x is the original value, u is the mean and s is the standard deviation.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x_train, x_test , y_train , y_test = train_test_split (x,y,random_state=40)
```

```python
x_train.head()
```

```python
scaler = StandardScaler().fit(x_train)
```

```python
scaler.mean_
```

```python
scaler.scale_ # variouns
```

```python
scaler.transform(x_train)
```

```python
x_train_scaled = scaler.transform(x_train)
```

```python
print(x_train_scaled)
```

After_transaformation

```python
print(x_train_scaled.mean(axis=0))
```

```python
print(x_train_scaled.std(axis=0))
```
